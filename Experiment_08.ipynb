```python
# EXPERIMENT-8: Develop program for Correlation and scatter plots

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic datasets
np.random.seed(42)
x = np.random.randint(50, 100, 100)  # Dataset 1: random integers between 50 and 100
y = x * 0.8 + np.random.randn(100) * 5  # Dataset 2: scaled x with some noise added

# Combine datasets into a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Compute the correlation between X and Y
correlation = data.corr().iloc[0, 1]  # Extract correlation coefficient
print("Correlation between X and Y:", correlation)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['Y'], color='blue', alpha=0.6, label='Data Points')

# Add title, labels, grid, and legend
plt.title("Scatter Plot with Correlation")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
