```python
# EXPERIMENT-7: Develop a Program for Normal Curves 

import numpy as np
import matplotlib.pyplot as plt

# For reproducibility, set a seed for the random number generator.
np.random.seed(0)

# Simulate a dataset of exam scores (real values) using a normal distribution.
# Let's assume the scores are roughly centered around 65 with a standard deviation of 15.
real_data = np.random.normal(65, 15, 500)

# Compute the sample mean and standard deviation from the dataset.
mu = np.mean(real_data)
sigma = np.std(real_data)

print("Calculated Mean:", mu)
print("Calculated Standard Deviation:", sigma)

# Plot a histogram of the real_data.
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(real_data, bins=30, density=True, alpha=0.6, color='g', label="Real Data Histogram")

# Compute the x-values for plotting the normal probability density function (PDF).
x = np.linspace(bins[0], bins[-1], 1000)
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (x - mu)**2 / (2 * sigma**2))

# Overlay the normal curve on top of the histogram.
plt.plot(x, pdf, 'k', linewidth=2, label=f'Normal Curve\n(mean = {mu:.2f}, sigma = {sigma:.2f})')

# Add title, labels, legend, and grid.
plt.title("Histogram of Exam Scores with Fitted Normal Curve")
plt.xlabel("Exam Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)

# Display the plot.
plt.show()
```
