```python
#EXPERIMENT-4: Develop a Program for Frequency Distributions

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
grades = ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'A', 'B', 'A', 'C', 'C', 'B', 'A']
grades_series = pd.Series(grades)
frequency_distribution = grades_series.value_counts().sort_index()  # Sorting for better visualization
print("Frequency Distribution:\n", frequency_distribution)
plt.bar(frequency_distribution.index, frequency_distribution.values, color='blue', alpha=0.7)
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.title("Bar Graph of Grade Distribution")
plt.show()
grade_mapping = {'A': 90, 'B': 80, 'C': 70}
numerical_scores = [grade_mapping[g] for g in grades]

plt.hist(numerical_scores, bins=3, color='green', alpha=0.7, edgecolor='black')
plt.xlabel("Numerical Score")
plt.ylabel("Frequency")
plt.title("Histogram of Numerical Grade Distribution")
plt.show()

frequency_distribution.plot(kind='bar', color='red', alpha=0.7)
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.title("Bar Plot of Grade Distribution (Pandas)")
plt.show()


class_intervals = list(grade_mapping.values())  # Using numerical scores as class intervals
freq_counts = [numerical_scores.count(interval) for interval in class_intervals]

plt.plot(class_intervals, freq_counts, marker='o', linestyle='-', color='purple', label="Frequency Polygon")
plt.xlabel("Numerical Grade")
plt.ylabel("Frequency")
plt.title("Frequency Polygon of Numerical Grade Distribution")
plt.legend()
plt.grid(True)
plt.show()
```
